```python
# import
import numpy as np
import pandas as pd
import json
```

```python
path = r"Data Pivot Party 19012021.xlsx"
df_xlsx = pd.ExcelFile(path)
```

```python
df_xlsx.sheet_names
```

```python
df = pd.read_excel(df_xlsx,'uc_2020_12')
df
```

```python
# data cleansing
df.drop(columns=['Unnamed: 0'], inplace=True) # hilangin kolom kosongan
df['Intent'] = df['Intent'].str.replace("usr.ask_", "") # hilangin usr.ask_
df['Message Request'] = df['Message Request'].apply(lambda x: json.loads(x)['msg'][0]['text']) # buka json, ambil text
df
```
